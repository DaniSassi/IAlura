```python
# BLOCO 3: Função para Simular a Análise de Sentimento
def responder(comentario):
    """Simula a análise de sentimento do comentário e retorna uma resposta apropriada com emojis."""
    print(f"Comentário recebido: {comentario}")
    score = 0.0  # Inicializa o score como neutro

    if "adorei" in comentario.lower() or "incrível" in comentario.lower() or "feliz" in comentario.lower():
        score = 0.8
        print(f"Sentimento simulado: Positivo (score={score:.2f})")
        return random.choice(biblioteca_respostas["positivo"])
    elif "triste" in comentario.lower() or "péssimo" in comentario.lower() or "não gostei" in comentario.lower():
        score = -0.7
        print(f"Sentimento simulado: Negativo (score={score:.2f})")
        return random.choice(biblioteca_respostas["negativo"])
    elif "haha" in comentario.lower() or "kkk" in comentario.lower() or "engraçado" in comentario.lower():
        score = 0.5
        print(f"Sentimento simulado: Humor (score={score:.2f})")
        return random.choice(biblioteca_respostas["humor"])
    elif "motivação" in comentario.lower() or "ânimo" in comentario.lower() or "força" in comentario.lower():
        score = 0.6
        print(f"Sentimento simulado: Motivação (score={score:.2f})")
        return random.choice(biblioteca_respostas["motivação"])
    else:
        score = 0.0
        print(f"Sentimento simulado: Neutro (score={score:.2f})")
        return random.choice(biblioteca_respostas["neutro"])
```
